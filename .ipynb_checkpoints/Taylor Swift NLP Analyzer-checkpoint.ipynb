```python
import requests
import json
from pprint import pprint
from bs4 import BeautifulSoup
import pandas as pd
import nltk.classify.util
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
file_content = open("all_tswift_lyrics.txt").read()
tokens = nltk.word_tokenize(file_content)
print(tokens)
```

```python
useful_words = [word for word in tokens if word not in stopwords.words('english')]
print(useful_words)
```

```python
freq_dist = nltk.FreqDist(useful_words)

freq_dist.most_common(25)
```
